# NO. 1 This is for cartesian coordiate system!

```python
import matplotlib.pyplot as plt
import numpy as np
x = []
y = []
m = 1
b = 1

for i  in range(9):
    z =  m * i + b
    x.append(i)
    y.append(z)

ax = plt.axes()
plt.plot(x,y,c = 'red',)
plt.xlim([0, 9])
plt.ylim([0,12])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.annotate('y = mx + b',(5,9))
ax.annotate('m = 1', (6,2))
ax.annotate('b = 1', (6,3))
plt.title('Cartesian Coordinate System')
ax.xaxis.grid(True, which = 'major', linestyle = 'dotted')
ax.yaxis.grid(True, which = 'major', linestyle = 'dotted')


```

# NO. 3 Upgraded Hough Space Transformation

```python
# upgraded hough space transformation--- hesse projection
# polar representation for line
# d = x*cos(theta) + y* sin(theta)
# This algorithm did not consider the vote!!
import math
```

```python
X = np.array(x)
Y = np.array(y)
n = len(x)
d = np.zeros(shape=(9,181))
# generate matrix for the calculation
for j in range(n):
    for i in range(181):
        theta = i * math.pi/180
        a = X[j] * math.cos(theta) + Y[j] * math.sin(theta)
        d[j,i] = a
        
# data visulization for the calculation
for j in range(n):
    plt.plot(range(181),d[j,:],'r')

ax = plt.axes()
plt.xlim([0, 180])
ax.set_xlabel('theta')
ax.set_ylabel('d')
ax.annotate('Point',xy=(140,2))
ax.xaxis.grid(True, which = 'major', linestyle = 'dotted')
ax.yaxis.grid(True, which = 'major', linestyle = 'dotted')
b = 'Polar Representation for Line'
plt.title(b)
```
